## Descriptive Statistics Overview
*Calculate central tendency and dispersion for sales_amount, units_sold, and customer_satisfaction to establish a performance baseline.*

## Sales Trend Analysis
*Analyze monthly and quarterly sales_amount trends to identify seasonal patterns and growth trajectories over the 2023-2024 period.*

```
CHART_PATH = 'chart_task_002.png'

import pandas as pd
import matplotlib.pyplot as plt

try:
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    df = df.sort_values('date')

    monthly_sales = df.groupby(df['date'].dt.to_period('M'))['sales_amount'].sum()
    quarterly_sales = df.groupby(['year', 'quarter'])['sales_amount'].sum()

    print("--- Sales Trend Analysis ---")
    print(f"Total Sales: ${df['sales_amount'].sum():,.2f}")
    print("\nMonthly Sales Trend:")
    print(monthly_sales)
    print("\nQuarterly Sales Performance:")
    print(quarterly_sales)

    plt.figure(figsize=(12, 6))
    monthly_sales.plot(kind='line', marker='o', linestyle='-', color='b')
    plt.title('Monthly Sales Trend (2023-2024)')
    plt.xlabel('Month')
    plt.ylabel('Total Sales Amount')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.savefig(CHART_PATH, bbox_inches='tight')
    plt.close('all')

    peak_month = monthly_sales.idxmax()
    print(f"\nPeak sales month: {peak_month} with ${monthly_sales.max():,.2f}")
    
    growth = (monthly_sales.iloc[-1] - monthly_sales.iloc[0]) / monthly_sales.iloc[0] * 100
    print(f"Overall growth trajectory (first vs last recorded month): {growth:.2f}%")

except Exception as e:
    print(f"Error during trend analysis: {e}")
```

**Insight:** Total sales peaked in May 2023 at $1,561,496, followed by a significant decline in performance. Comparing January 2023 to May 2024, the company experienced a 47.79% decrease in monthly sales volume. Quarterly data confirms this downward trajectory, with Q2 2024 sales reaching only $1,613,623 compared to the $3,038,711 achieved in Q2 2023.

*Confidence: 95%*

## Correlation Matrix of Performance Metrics
*Examine relationships between sales_amount, units_sold, sales_target, and customer_satisfaction to determine if higher satisfaction correlates with higher sales.*

```
CHART_PATH = 'chart_task_003.png'

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

try:
    cols_to_correlate = ['sales_amount', 'units_sold', 'sales_target', 'customer_satisfaction']
    corr_matrix = df[cols_to_correlate].corr()

    print("Correlation Matrix:")
    print(corr_matrix)

    sales_sat_corr = corr_matrix.loc['sales_amount', 'customer_satisfaction']
    print(f"\nKey Finding: Correlation between Sales Amount and Customer Satisfaction: {sales_sat_corr:.3f}")

    if sales_sat_corr > 0:
        print("Observation: There is a positive correlation between customer satisfaction and sales.")
    elif sales_sat_corr < 0:
        print("Observation: There is a negative correlation between customer satisfaction and sales.")
    else:
        print("Observation: There is no linear correlation between customer satisfaction and sales.")

    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Correlation Matrix of Performance Metrics')
    plt.savefig(CHART_PATH, bbox_inches='tight')
    plt.close('all')

except Exception as e:
    print(f"Error during correlation analysis: {e}")
```

**Insight:** The analysis reveals a near-zero correlation (-0.002) between customer satisfaction scores and total sales amount, indicating that current satisfaction levels have no measurable impact on revenue generation. Conversely, sales amount shows a strong positive correlation with sales targets (0.917) and units sold (0.843), confirming that volume and quota-setting are the primary drivers of performance.

## Regional and Product Performance Comparison
*Compare average sales_amount and return rates across different regions and products to identify high-performing segments versus underperforming ones.*

```
CHART_PATH = 'chart_task_004.png'

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    df['return_rate'] = (df['returns'] / df['units_sold']) * 100
    
    region_perf = df.groupby('region')[['sales_amount', 'return_rate']].mean().sort_values('sales_amount', ascending=False)
    product_perf = df.groupby('product')[['sales_amount', 'return_rate']].mean().sort_values('sales_amount', ascending=False)
    
    print("--- Regional Performance Summary ---")
    print(region_perf)
    print("\n--- Product Performance Summary ---")
    print(product_perf)
    
    top_region = region_perf.index[0]
    top_product = product_perf.index[0]
    print(f"\nTop performing region by average sales: {top_region}")
    print(f"Top performing product by average sales: {top_product}")

    fig, axes = plt.subplots(2, 1, figsize=(10, 12))
    
    sns.barplot(x=region_perf.index, y=region_perf['sales_amount'], ax=axes[0], palette='viridis')
    axes[0].set_title('Average Sales Amount by Region')
    axes[0].set_ylabel('Avg Sales ($)')
    
    sns.barplot(x=product_perf.index, y=product_perf['sales_amount'], ax=axes[1], palette='magma')
    axes[1].set_title('Average Sales Amount by Product')
    axes[1].set_ylabel('Avg Sales ($)')
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.savefig(CHART_PATH, bbox_inches='tight')
    plt.close('all')

except Exception as e:
    print(f"Error during analysis: {e}")
```

**Insight:** The North region leads with the highest average sales of $52,751, while the South region trails at $40,082. Product C is the clear volume leader, generating an average of $71,981 per transaction, though it also carries the highest return rate at 4.73%. Conversely, Product D generates the lowest average sales at $25,648 but maintains the best return performance at 3.26%.

*Confidence: 95%*

## Outlier Detection in Returns and Units Sold
*Identify specific records where units_sold or returns deviate significantly from the norm to investigate potential data entry errors or anomalous sales events.*

```
CHART_PATH = 'chart_task_005.png'

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

try:
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    
    def get_outliers(series):
        q1 = series.quantile(0.25)
        q3 = series.quantile(0.75)
        iqr = q3 - q1
        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr
        return (series < lower) | (series > upper)

    df['outlier_units'] = get_outliers(df['units_sold'])
    df['outlier_returns'] = get_outliers(df['returns'])
    
    outlier_df = df[df['outlier_units'] | df['outlier_returns']]
    
    print("--- Outlier Detection Summary ---")
    print(f"Total records identified as outliers: {len(outlier_df)}")
    print(f"Outliers in Units Sold: {df['outlier_units'].sum()}")
    print(f"Outliers in Returns: {df['outlier_returns'].sum()}")
    
    if not outlier_df.empty:
        print("\nSample of anomalous records:")
        print(outlier_df[['date', 'product', 'units_sold', 'returns', 'sales_amount']].head(10).to_string())
    
    plt.figure(figsize=(10, 6))
    plt.scatter(df['units_sold'], df['returns'], c=df['outlier_units'] | df['outlier_returns'], cmap='viridis', alpha=0.6)
    plt.xlabel('Units Sold')
    plt.ylabel('Returns')
    plt.title('Outlier Detection: Units Sold vs Returns')
    plt.colorbar(label='Is Outlier')
    plt.grid(True, linestyle='--', alpha=0.7)
    
    plt.savefig(CHART_PATH, bbox_inches='tight')
    plt.close('all')

except Exception as e:
    print(f"Error during outlier analysis: {e}")
```

**Insight:** The analysis identified 19 anomalous records, with a notable concentration of 15 outliers in return volumes and 10 in units sold. Specifically, Product C consistently appears in the outlier list, with return counts reaching as high as 26 units and sales volumes peaking at 384 units in a single transaction.

*Confidence: 85%*
